```python
#Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Read CSV and combine files
city_df = pd.read_csv("raw_data/city_data.csv")
ride_df = pd.read_csv("raw_data/ride_data.csv")
combined_df = pd.merge(city_df, ride_df, on ="city")
combined_df.head()
```

```python
#Find average fare/city
avg_fare = combined_df.groupby(["city"]).mean()["fare"]
#Find # of Rides/city
num_rides = combined_df.groupby(["city"]).count()["ride_id"]
#Find # of Driver/city
num_driver = combined_df.groupby(["city"]).mean()["driver_count"]
##Reformat num_driver
driver_count=num_driver.map('{:.0f}'.format)

#Find City Type
type_group = city_df.groupby(["type"])
city_type = list(type_group.groups)#['Rural', 'Suburban', 'Urban']

#Find City Type column
city_df2=city_df.drop_duplicates(["city"])
type_col=city_df2["type"]
city_col=city_df2["city"]
type_df=pd.DataFrame({"city": city_col,"City Type": type_col})
```

```python
type_change= pd.to_numeric(type_col, errors='ignore')
```

```python
#Group average fare, number of rides and driver count per city
df2=pd.DataFrame({"Average Fare per City": avg_fare,
                 "Number of Rides": num_rides,
                 "Number of Driver":num_driver})
df2.reset_index(level=0, inplace=True)
df3=pd.merge(type_df,df2,on="city")
df3.head()
#df2.head()
```

```python
#Plot using seaborn
import seaborn as sns

# Bubble plot

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
colors = {'Urban':'gold', 'Rural':'lightskyblue', 'Suburban':'lightcoral'}

ax.scatter(x=df3["Number of Rides"],y=df3["Average Fare per City"], s=df3["Number of Driver"]*10, 
           c=df3["City Type"].apply(lambda x: colors[x]), edgecolors="blue", linewidth=2, label =colors)


# Add titles (main and on axis) 
plt.xlabel("Number of Rides per City") 
plt.ylabel("Average Fare per City") 
plt.suptitle("Pyber Ride Sharing Data")
#plt.title("Circle size correlated to driver count")
fig.text(.03,0,"Note: Circle size correlated to driver count")
fig.text(.03,.88,"Color Legend: Light Sky Blue = Rural, Light Coral = Suburban, Gold=Urban")
plt.legend(colors)
plt.show()
fig.clear()
```

```python
#Observable Trend 1: There're more driver in Urban than in Rural or Suburban
#Observable Trend 2: There're higher # of rides per city from Urban than in Rural or Suburban
#Observable Trend 3: Average fare per city for Urban is the lowest.
```

```python
#Find Total Fare
total_fare=combined_df["fare"].sum()
total_fare
```

```python
type_fare=combined_df.groupby(["type"]).sum()["fare"]
percent_type=type_fare/total_fare
percent_type
pie_fare_df=pd.DataFrame({"Percent of Total Fares":percent_type})
pie_fare_df.reset_index(level=0, inplace=True)
#plot
fig.clear()
plt.figure(figsize=(12,6))
# plot chart
ax1 = plt.subplot(121)
pie_fare_df.plot(kind='pie', y = 'Percent of Total Fares', ax=ax1, autopct='%1.1f%%', 
 startangle=200, shadow=False, labels=pie_fare_df['type'], legend = False, fontsize=14)
plt.suptitle("Percent of Total Fares by City Type")
plt.show()
```

```python
#Find Total Rides
total_rides=combined_df["ride_id"].sum()
type_rides=combined_df.groupby(["type"]).sum()["ride_id"]
percent_rides=type_rides/total_rides
pie_rides_df=pd.DataFrame({"Percent of Total Rides":percent_rides})
pie_rides_df.reset_index(level=0, inplace=True)
#plot
fig.clear()
plt.figure(figsize=(12,6))
# plot chart
ax1 = plt.subplot(121)
pie_rides_df.plot(kind='pie', y = "Percent of Total Rides", ax=ax1, autopct='%1.1f%%', 
 startangle=200, shadow=False, labels=pie_rides_df['type'], legend = False, fontsize=14)
plt.suptitle("Percent of Total Rides by City Type")
plt.show()
```

```python
#Find Total Rides
total_drivers=combined_df["driver_count"].sum()
type_drivers=combined_df.groupby(["type"]).sum()["driver_count"]
percent_drivers=type_drivers/total_drivers
pie_drivers_df=pd.DataFrame({"Percent of Total Drivers":percent_drivers})
pie_drivers_df.reset_index(level=0, inplace=True)
#plot
fig.clear()
plt.figure(figsize=(12,6))
# plot chart
ax1 = plt.subplot(121)
pie_drivers_df.plot(kind='pie', y = "Percent of Total Drivers", ax=ax1, autopct='%1.1f%%', 
 startangle=90, shadow=False, labels=pie_drivers_df['type'], legend = False, fontsize=12)
plt.suptitle("Percent of Total Drivers by City Type")
plt.show()
```

```python
#Observable Trend 1: There're more driver in Urban than in Rural or Suburban
#Observable Trend 2: There're higher # of rides per city from Urban than in Rural or Suburban
#Observable Trend 3: Average fare per city for Urban is the lowest.
```
